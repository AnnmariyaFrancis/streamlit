```python
import cv2

# Load the image
image = cv2.imread(r'D:\DL_Projects\helmet_detection\New folder (2)\without_helmet\171.jpg')

# Load the license plate Haar Cascade classifier (you can download one or create your own)
plate_cascade = cv2.CascadeClassifier('D:\DL_Projects\helmet_detection\haarcascade_license_plate_rus_16stages.xml')

# Convert the image to grayscale (required for object detection)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect license plates in the image
plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(50, 25))

# Loop through the detected plates (there may be multiple)
for (x, y, w, h) in plates:
    # Draw a rectangle around the detected license plate
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Extract the detected license plate from the image
    plate_roi = image[y:y + h, x:x + w]
else:
    print("no plate is identified")
# Save or display the extracted license plate
cv2.imwrite('extracted_license_plate.jpg', plate_roi)
cv2.imshow('Extracted License Plate', plate_roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
